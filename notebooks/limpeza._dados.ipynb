```python
#COLUNA 0
#VALORES ESPERADOS
# Definindo os valores esperados na coluna
valores_esperados = ['ACIDENTE', 'INCIDENTE GRAVE', 'INCIDENTE']

# Criando um filtro para linhas que não correspondem aos valores esperados
filtro_diferentes = ~df['ocorrencia_classificacao'].isin(valores_esperados)
linhas_diferentes = df[filtro_diferentes]

# Contando quantos valores diferentes existem
quantidade_diferentes = filtro_diferentes.sum()

# Imprimindo a quantidade e, se houver, quais são as linhas
print(f'Número de valores diferentes do esperado na coluna ocorrencia_classificacao: {quantidade_diferentes}')
if quantidade_diferentes > 0:
    print("Linhas com valores diferentes do esperado na coluna ocorrencia_classificacao:")
    print(linhas_diferentes[['ocorrencia_classificacao']])
else:
    print("Todos os valores na coluna ocorrencia_classificacao são 'ACIDENTE', 'INCIDENTE' ou 'INCIDENTE GRAVE'")
```

```python
# Função para converter e ajustar os valores de latitude e longitude
def ajustar_valores(valor):
    if isinstance(valor, str) and valor.strip().lower() == 'Não Informado':
        return valor
    try:
        valor_float = float(str(valor).replace(',', '.').replace(' ', ''))
        return valor_float / 1000000000000 if abs(valor_float) > 1000000000000 else valor_float
    except ValueError:
        return valor

# Aplicar a função nas colunas de latitude e longitude e criar novas colunas
df['lat_corrigida'] = df['ocorrencia_latitude'].apply(ajustar_valores)
df['long_corrigida'] = df['ocorrencia_longitude'].apply(ajustar_valores)

# Verificar as mudanças aplicadas
print("Valores originais e ajustados de latitude:")
print(df[['ocorrencia_latitude', 'lat_corrigida']].head(10))
print("Valores originais e ajustados de longitude:")
print(df[['ocorrencia_longitude', 'long_corrigida']].head(10))

# Salvar o DataFrame com as alterações realizadas
caminho_arquivo_salvo = 'C:/Users/USER/Desktop/desafio_jcpoli/data/processed/acidentes_processados.xlsx'
df.to_excel(caminho_arquivo_salvo, index=False)

```

```python
# Expressão regular para verificar o formato "hh:mm:ss"
padrao_hora = r'^\d{2}:\d{2}:\d{2}$'

# Aplicar a expressão regular para encontrar não correspondências
mascara_formato_incorreto = ~df['ocorrencia_hora'].astype(str).str.match(padrao_hora)
horas_incorretas = df[mascara_formato_incorreto]

# Verificar se existem valores incorretos e exibi-los
if not horas_incorretas.empty:
    print("Entradas com formato de hora incorreto:")
    print(horas_incorretas['ocorrencia_hora'])
else:
    print("Todas as entradas em 'ocorrencia_hora' estão no formato correto 'hh:mm:ss'.")

# Salvar o DataFrame atualizado no diretório processed
caminho_arquivo_processado = r'C:\Users\USER\Desktop\desafio_jcpoli\data\processed\acidentes_processados.xlsx'
df.to_excel(caminho_arquivo_processado, index=False)
```

```python
# Lista de colunas para padronizar
colunas = ['investigacao_aeronave_liberada', 'ocorrencia_saida_pista']

# Padronizar "sim" e "não"
for coluna in colunas:
    df[coluna] = df[coluna].str.strip().str.upper()
    df[coluna] = df[coluna].replace(['sim', 'Sim', 'SIM', 's', 'S'], 'SIM')
    df[coluna] = df[coluna].replace(['Não', 'Nao', 'nao', 'nâo', 'n', 'ñ','N','Ñ'], 'NÃO')

# Verificar se existem valores fora de 'SIM' e 'NÃO'
for coluna in colunas:
    valores_fora = df[~df[coluna].isin(['SIM', 'NÃO'])][coluna].unique()
    if len(valores_fora) > 0:
        print(f"Valores fora de 'SIM' e 'NÃO' em {coluna}: {valores_fora}")
    else:
        print(f"Todas as entradas em {coluna} estão corretamente padronizadas como 'SIM' ou 'NÃO'.")

# Salvar o DataFrame atualizado no diretório processed
caminho_arquivo_processado = r'C:\Users\USER\Desktop\desafio_jcpoli\data\processed\acidentes_processados.xlsx'
df.to_excel(caminho_arquivo_processado, index=False)
```

```python
# Função para remover o último dígito de um ano
def remover_ultimo_digito(ano):
    if isinstance(ano, str) and ano.isdigit():
        return int(ano[:-1])
    elif isinstance(ano, (int, float)) and not pd.isna(ano):
        return int(ano // 10)
    return ano

# Aplicar a função na coluna 'aeronave_ano_fabricacao'
df['aeronave_ano_fabricacao'] = df['aeronave_ano_fabricacao'].apply(remover_ultimo_digito)

# Verificar as primeiras linhas para confirmar a mudança
print("Após a correção:")
print(df['aeronave_ano_fabricacao'].head())

# Salvar o DataFrame atualizado no diretório processed
caminho_arquivo_processado = r'C:\Users\USER\Desktop\desafio_jcpoli\data\processed\acidentes_processados.xlsx'
df.to_excel(caminho_arquivo_processado, index=False)
```
