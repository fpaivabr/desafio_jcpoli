```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_excel('acidentes_processados.xlsx')
```

```python
base.head()
```

```python
base.replace('Não Informado', pd.NA, inplace=True)

# Remover linhas com NaNs na coluna de tipo de operação da aeronave
base.dropna(subset=['aeronave_tipo_operacao'], inplace=True)

# Contar o número de ocorrências por tipo de operação da aeronave
contagem_por_tipo_operacao = base['aeronave_tipo_operacao'].value_counts()
```

```python
# Criar o gráfico de barras com largura ajustada
plt.figure(figsize=(10, 6))
contagem_por_tipo_operacao.plot(kind='bar', width=0.8, color='skyblue')  # Ajuste a largura conforme necessário
plt.title('Contagem do Tipo de Operação Realizada')
plt.xlabel('Tipo de Operação da Aeronave')
plt.ylabel('Contagem de Ocorrências')
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotacionar e ajustar tamanho da fonte dos rótulos
plt.tight_layout()  # Ajustar layout para evitar corte de rótulos
plt.show()
```
