# Abertura de arquivo

Começando com a importação do pandas para abrir os dados

```python
import pandas as pd
```

Leitura dos dados

```python
df = pd.read_excel('/content/drive/MyDrive/Desafio de dados VI/acidentesTransito_cenipa_2010a2021.xlsx')
```

```python
df.info()
```

# Tratamento de dados

```python
df['aeronave_ano_fabricacao'] = df['aeronave_ano_fabricacao'] // 10
```

```python
print(len(df))
```

Separando as linhas para ter acidentes unicos, sem repetições

```python
df = df.drop_duplicates(subset=['divulgacao_relatorio_numero'])
```

Quantas linhas existem agora

```python
print(len(df))
```

Remover colunas desnecessárias

```python
df = df.drop(columns=['ocorrencia_latitude', 'ocorrencia_longitude', 'ocorrencia_aerodromo', 'ocorrencia_dia', 'ocorrencia_pais', 'fator_nome', 'fator_aspecto', 'fator_condicionante', 'fator_area', 'investigacao_status', 'investigacao_aeronave_liberada', 'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao', 'taxonomia_tipo_icao', 'recomendacao_numero', 'recomendacao_dia_assinatura', 'recomendacao_dia_encaminhamento', 'recomendacao_dia_feedback', 'recomendacao_destinatario'])
```

```python
df.info()
```

# Criação de Gráficos

Importação da biblioteca para criação de gráficos

```python
import matplotlib.pyplot as plt
```

Criação de gráficos com base nos dados obtidos

```python
plt.figure(figsize=(5, 6))  # Define o tamanho da figura
df['ocorrencia_classificacao'].value_counts().plot(kind='bar', color='skyblue')  # Cria o gráfico de barras
plt.title('Tipos de ocorrência')  # Define o título do gráfico
plt.xlabel('Ocorrências')  # Define o rótulo do eixo x
plt.ylabel('Nº casos')  # Define o rótulo do eixo y
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adiciona linhas de grade no eixo y
plt.show()
```
