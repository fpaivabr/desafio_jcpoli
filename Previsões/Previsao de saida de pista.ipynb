```python
ct.set_output(transform="pandas")
```

```python
#transformando as colunas
base_transformada = ct.fit_transform(base)
base_transformada.rename(columns ={"ordinalencoder__ocorrencia_saida_pista":"ocorrencia_saida_pista"}, inplace=True)
```
